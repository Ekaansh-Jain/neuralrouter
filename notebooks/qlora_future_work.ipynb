# QLoRA Router Distillation — Future Work

**Status: experiment / documented future work — NOT in the live serving path.**

The production classifier is a CPU-servable TF-IDF + LogisticRegression model
(see `backend/fine_tune/train_encoder.py`). This notebook sketches the more
ambitious idea of distilling the evaluator's routing judgments into a
QLoRA-fine-tuned Llama 3 8B.

### Why it is not live
There is no free way to *serve* custom 8B weights with acceptable latency
(Groq/OpenRouter only serve their own models). A complexity label is a short
text-classification task that does not need a generative model, so the encoder
approach is both cheaper and actually deployable.

### Sketch
1. Export labeled data with `fine_tune/export_dataset.py` (excludes fallback rows).
2. Load 4-bit Llama 3 8B with Unsloth on Colab.
3. QLoRA fine-tune on (query → complexity) pairs.
4. Evaluate; if it beats the TF-IDF baseline AND a serving option exists,
   consider promoting it behind the `Classifier` protocol.
